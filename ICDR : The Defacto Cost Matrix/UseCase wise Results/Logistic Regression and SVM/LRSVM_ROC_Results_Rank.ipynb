```python
final_data = pd.concat([data,data2])
ALGO = ["CSDT"]*4+["CSLR"]*4+["CSRF"]*4+["CSSVM"]*4+["CSXG"]*4
ALGO=ALGO*6
algo = ["DT"]*4+["LR"]*4+["RF"]*4+["SVM"]*4+["XG"]*4
algo = algo*6
final_data["ALGO"]=ALGO+algo
final_data["ALGORITHM"] = ["CS"]*120+["NONCS"]*120
final_data
```
