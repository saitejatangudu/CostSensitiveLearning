```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00383/risk_factors_cervical_cancer.csv",dtype=object)
data.head()
```

```python
data.dtypes
```

```python
Null_data = pd.DataFrame()
for col in data.columns:
    Null_data[col] = pd.to_numeric(data[col], errors='coerce').convert_dtypes() 
Null_data.head()    
```

```python
data = Null_data
#data.fillna(0,inplace=True)
data.head()
```

```python
from sklearn.impute import KNNImputer 
```

```python
Imputer = KNNImputer(n_neighbors=5)
Impute_data= pd.DataFrame(Imputer.fit_transform(data))
for i,j in enumerate(data.columns):
    data[j] = Impute_data[i] 
data.head()
```

```python
# import module
from sklearn.preprocessing import MinMaxScaler
# scale features
scaler = MinMaxScaler()
model=scaler.fit(data)
Impute_data=model.transform(data)

# print scaled features
data = pd.DataFrame(Impute_data,columns = data.columns)
data.head()
```

```python
data_stack = data[["Hinselmann","Schiller","Citology","Biopsy"]].stack()
data_stack.head()
```

```python
data["class"] = pd.Series(pd.Categorical(data_stack[data_stack!=0].index.get_level_values(1)))
```

```python
data.drop(["Hinselmann","Schiller","Citology","Biopsy"],axis=1,inplace=True)
data.head()
```

```python
from collections import Counter
Counter(data["class"])
```

```python
data["class"] = np.where(data["class"]=='Biopsy',1,0)
```

### DATA RESAMPLING

### GENERATING TEST DATA

```python
X = data.drop(['class'],axis=1)
y = data["class"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
print(sum(y_train==1)
,sum(y_train==0)
,sum(y_test==1)
,sum(y_test==0)
     )
```

```python
x_test.to_csv("cervical_test_data.csv",index=False)
y_test.to_csv("cervical_test_label.csv",index=False)
```

### GENERATING EXT_IMB DATASET(05:95)

```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import ADASYN 
from imblearn.under_sampling import RandomUnderSampler
```

```python
X, y = x_train,y_train
print('Original dataset shape %s' % Counter(y))
```

```python
no_minority, no_majority = (Counter(y)[1]+Counter(y)[0])*0.05,(Counter(y)[1]+Counter(y)[0])*0.95
rnd = RandomUnderSampler(random_state=42,sampling_strategy={1:round(no_minority),0:Counter(y)[0]})
X_final, y_final = rnd.fit_resample(X, y)
print('Resampled dataset shape with 5:95 ratio%s' % Counter(y_final))
```

```python
X_final.to_csv("cervical_ext_imb_data.csv",index=False)
y_final.to_csv("cervical_ext_imb_label.csv",index=False)
```

### GENERATING 15285_IMB DATASET(15:85)

```python
X, y = x_train,y_train
print('Original dataset shape %s' % Counter(y))
```

```python
no_minority, no_majority = (Counter(y)[1]+Counter(y)[0])*0.15,(Counter(y)[1]+Counter(y)[0])*0.85
ada = ADASYN(random_state=42,sampling_strategy={0:Counter(y)[0],1:round(no_minority)})
X_res, y_res = ada.fit_resample(X, y)
print('Resampled dataset shape After oversampling minority class %s' % Counter(y_res))
rnd = RandomUnderSampler(random_state=42,sampling_strategy={0:round(no_majority),1:Counter(y_res)[1]})
X_final, y_final = rnd.fit_resample(X_res, y_res)
print('Resampled dataset shape with 15:85 ratio%s' % Counter(y_final))
```

```python
X_final.to_csv("cervical_15285_imb_data.csv",index=False)
y_final.to_csv("cervical_15285_imb_label.csv",index=False)
```

```python
X_final.to_csv("cervical_mod_imb_data.csv",index=False)
y_final.to_csv("cervical_mod_imb_label.csv",index=False)
```
