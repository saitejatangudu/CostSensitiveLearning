```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Example data (replace with your actual data)
data = new_df

# Get unique metric names and methods
metric_names = data['MetricName'].unique()
methods = data['Method'].unique()

# Calculate the number of rows and columns for the grid
num_rows = len(metric_names) // 2 + len(metric_names) % 2  # Ensures at least one row per metric
num_cols = 2

# Create a color palette with distinct colors for each method
color_palette = sns.color_palette("Set2", len(methods))

# Create a grid of line charts
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 4 * num_rows))

# Flatten the axs array to iterate over it
axs = axs.flatten()

# Iterate over each metric
for i, metric in enumerate(metric_names):
    metric_data = data[data['MetricName'] == metric]
    
    # Calculate the average value for each method within the metric
    avg_values = metric_data.groupby('Method')['Value'].mean()
    
    # Identify the method with the highest average value
    best_method = avg_values.idxmax()
    
    # Plot lines for each method
    for j, method in enumerate(methods):
        method_data = metric_data[metric_data['Method'] == method]
        
        # Determine the subplot index
        subplot_idx = i % num_rows * num_cols + i // num_rows
        
        # Assign a larger linewidth to the best method
        linewidth = 5 if method == best_method else 1.5
        
        # Plot line in the corresponding subplot with a specific color and linewidth
        axs[subplot_idx].plot(method_data['DatasetName'], method_data['Value'], marker='o',
                              color=color_palette[j], linewidth=linewidth, label=method)
        
        # Set plot title and labels
        axs[subplot_idx].set_title(metric)
        axs[subplot_idx].set_ylabel('Value')
        axs[subplot_idx].legend()

# Remove any empty subplots
if len(metric_names) < num_rows * num_cols:
    for empty_subplot in axs[len(metric_names):]:
        empty_subplot.remove()

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
