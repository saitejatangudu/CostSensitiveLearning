```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
from matplotlib.transforms import Bbox
from ds_pipe.datasets.dataset_loader import DatasetCollections
from ds_pipe.datasets.dataset_metadata import DatasetInformation

print("Imports loaded")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score,recall_score, silhouette_score
```

```python
def n_fold_cross_validation(dataset, n, n_neighbors=10): 
    kf = KFold(n_splits = n)
    kNN_accuracies = []
    kNN_recall_scores = []

    ffnn_accs,ffnn_recs = [],[]
    for train_index,test_index in kf.split(dataset.data): 
        X_train, X_test = dataset.data[train_index], dataset.data[test_index]
        y_train, y_test = dataset.target[train_index], dataset.target[test_index]
        
        kNN_clf = KNeighborsClassifier(n_neighbors=n_neighbors) 
        kNN_clf.fit(X=X_train, y=y_train)
        y_pred = kNN_clf.predict(X_test)
        kNN_accuracies.append(accuracy_score(y_test, y_pred))
        kNN_recall_scores.append(recall_score(y_test, y_pred, average="macro"))

        ffnn_clf = MLPClassifier(solver="lbfgs", alpha=1e-5, hidden_layer_sizes=(5,2)) 
        ffnn_clf.fit(X=X_train, y=y_train)
        y_pred = ffnn_clf.predict(X_test)
        ffnn_accs.append(accuracy_score(y_test, y_pred))
        ffnn_recs.append(recall_score(y_test, y_pred, average="macro"))

    result_dict = {'knn_acc': kNN_accuracies, 
                    "knn_rec": kNN_recall_scores, 
                    'ffnn_acc': ffnn_accs, 
                    "ffnn_rec": ffnn_recs}
    return result_dict

```

```python
import warnings 
warnings.simplefilter("ignore")
dc = DatasetCollections()
data_dict = dc.get_keel_dictionary()
data_dict.update(dc.get_chapelle_dictionary())
min_ir = 2
datasets = []
dim = []
ir = []
sizes = []
num_classes = []
imbalance_degree = []
lrids = []
imb_deg = []
imbalance_complexities = []
nies = []
capir = []
kNN_accuracies = []
kNN_recall_scores = []
ffnn_accuracies = []
ffnn_recall_scores = []
sil_scores = []

for dataset_name in data_dict.keys():
    dataset = data_dict[dataset_name]
    if len(dataset.target) < 10000: 
        di = DatasetInformation(dataset)
        if di.ir >= min_ir:
            datasets.append(dataset_name)
            dim.append(di.dimensionality)
            ir.append(di.ir)
            capir.append(di.complexity_adjusted_polytomous_imbalance_ratio())
            sizes.append(di.n_points)
            num_classes.append(di.num_classes)

            #imbalance_degree.append(di.imbalance_degree)
            nies.append(di.normalized_imb_ent)
            imbalance_complexities.append(2*di.imbalance_complexity)
            lrids.append(di.lrid_score)
            imb_deg.append(di.imbalance_degree)
            result_dict = n_fold_cross_validation(dataset,10,n_neighbors=10)

            kNN_accuracies.append(np.mean(result_dict['knn_acc']))
            kNN_recall_scores.append(np.mean(result_dict['knn_rec']))
            
            ffnn_accuracies.append(np.mean(result_dict['ffnn_acc']))
            ffnn_recall_scores.append(np.mean(result_dict['ffnn_rec']))
            sil_scores.append(silhouette_score(dataset.data, dataset.target))
        

    
```

```python
df = pd.DataFrame({"Datasets": datasets, "IR": ir, "Size": sizes, "Number of classes": num_classes, "LRID": lrids,"CapIR":capir, "NIE": nies, "Imbalance Complexity": imbalance_complexities,"Imbalance Degree": imb_deg, "knn_rec": kNN_recall_scores, "knn_acc": kNN_accuracies, "ffnn_rec": ffnn_recall_scores, "ffnn_acc": ffnn_accuracies, "dim": dim, "Silhouette": sil_scores})
```
